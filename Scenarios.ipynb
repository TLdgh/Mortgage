# Scenario Analysis

#### Suppose given the interest rate and duration of the loan, the original monthly payment is first calculated, and then the time to payout is re-estimated when the interest rate changes in the 37th month and an extra payment equal to 8% of the outstanding balance converted to monthly. 

```python
#利率不变
result1=Amortization(Balance=[200000], Rate=[[1, 0.0509/12]], N=25*12)
result1.CalcPayment(N=result1.N)

#利率3年后大幅上涨，没有提前还款，压力巨大
newPay=[1179.691]*36+[3159.245]*36+[1939.687]*228
result2=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.2/12], [73, 0.11/12]], Payment=newPay)
result2.CalcTime()

#利率3年后大幅上涨，后期提前还款，压力仍然巨大
newPay=[1179.691]*36+[3159.245]*36+[1939.687*2]*200
result3=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.2/12], [73, 0.11/12]], Payment=newPay)
result3.CalcTime()

#利率3年后大幅上涨，提前还款并且保持
newPay=[1179.691]*6+[1179.691*2]*160
result4=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.2/12], [73, 0.11/12]], Payment=newPay, BalancePctPay=0.08/12)
result4.CalcTime()

#利率3年后大幅上涨，提前还款并且支撑5年，后期还原
newPay=[1179.691]*6+[1179.691*2]*60+[1179.691]*200
result5=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.2/12], [73, 0.11/12]], Payment=newPay, BalancePctPay=0.08/12)
result5.CalcTime()

plotResult({"df1":result1.data, "df2":result2.data, "df3":result3.data, "df4":result4.data, "df5":result5.data})

#结论
#4 和 5利息差不多，但5后期还原，所以5最好
```

```python
#利率小幅下降，3年内提前还款，后期还原
newPay=[1179.691]*6+[1179.691*2]*36+[1179.691]*100
result6=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.04/12]], Payment=newPay, BalancePctPay=0.08/12)
result6.CalcTime()

#利率小幅下降，3年内提前还款，后期还原，没有 Balance Percentage Pay
newPay=[1179.691]*6+[1179.691*2]*36+[1179.691]*200
result7=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.04/12]], Payment=newPay)
result7.CalcTime()

#利率小幅下降，3年内提前还款并且保持
newPay=[1179.691]*6+[1179.691*2]*100
result8=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.04/12]], Payment=newPay, BalancePctPay=0.08/12)
result8.CalcTime()

#利率小幅下降，3年内提前还款并且保持，没有 Balance Percentage Pay
newPay=[1179.691]*6+[1179.691*2]*110
result9=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [37, 0.04/12]], Payment=newPay)
result9.CalcTime()

plotResult({"df6":result6.data, "df7":result7.data, "df8":result8.data, "df9":result9.data})

#结论：
#6 和 8 利息差不多，同时6后期还原，所以6最好
```
