```python
#Calculate Payment amount if n is known and a fixed interest rate:
result=Amortization(Balance=[30000], Rate=[[1, 0.03/12]])
result.CalcPayment(result.N) #use N=25 default value
```

```python
#Calculate Payment amount if variable interest rate:
result=Amortization(Balance=[30000], Rate=[[1, 0.05], [10, 0.03], [15, 0.02], [20, 0.01]], N=28)
result.CalcPayment(result.N) #use specified N value
```

```python
#Estimate time to payout gien a fixed rate and equal payment 
result=Amortization(Balance=[30000], Rate=[[1, 0.03]], Payment=[1964.03]*26)
result.CalcTime()
```

```python
#Estimate time to payout gien a fixed rate and variable payment 
result=Amortization(Balance=[30000], Rate=[[1, 0.03]], Payment=[1964.03]*12+[1000]*26)
result.CalcTime()
```

```python
#Estimate time to payout gien a variable rate and equal payment 
result=Amortization(Balance=[30000], Rate=[[1, 0.01], [13, 0.006], [22, 0.002]], Payment=[963]*30)
result.CalcTime()
```

```python
#Estimate time need to payout gien a variable rate and variable payment 
result=Amortization(Balance=[30000], Rate=[[1, 0.01], [13, 0.006]], Payment=[664.03]*12+[10000]+[1200]*12)
result.CalcTime()
```

# Scenario Analysis

#### Suppose given the interest rate and duration of the loan, the original monthly payment is first calculated, and then the time to payout is re-estimated when the interest rate changes in the 83th month and another lump sum payment of $20000 was made in the 121st month. 

```python
result1=Amortization(Balance=[200000], Rate=[[1, 0.0509/12]], N=25*12)
result1.CalcPayment(N=result1.N)

newPay=[1179.69]*6+[1179.69*2]*200
result2=Amortization(Balance=[200000], Rate=[[1, 0.0509/12], [12, 0.045/12]], Payment=newPay, BalancePctPay=0.08/12)
result2.CalcTime()

plotResult({"df1":result1.data, "df2":result2.data})
```
